figure.figsize changes the size of the graph

```python
import pandas as pd
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = [10, 5]
```

* Reads out Honey.csv
* Removes all ","
* Turns now valid numbers into actual floats.
* Drops any none float values. (cleans data)

```python
rawData = pd.read_csv("./Honey.csv")

rawData["Value"]=rawData["Value"].str.replace(",","")
rawData["Value"]=pd.to_numeric(rawData["Value"],errors="coerce")
rawData.dropna(subset=["Value"],inplace=True)
```

* Creates uniStates variable with all of the unique states in the data.
* Loops through each state.
* Grabs data based on state along with its Year and Value.

```python
uniStates = rawData.State.unique()

Large = 25000000 / 2
Larger = Large / 2
Medium = Larger / 2

for state in uniStates:
    honey = rawData[rawData["State"]==state].groupby("Year")["Value"]
    ssum = honey.sum()
    if ssum.sum() > Large:
        plt.plot(ssum.keys(),ssum,label=state)
plt.ticklabel_format(style='plain')
plt.xlabel("Year")
plt.ylabel("Production")
plt.title("Massive Producers")
plt.legend(bbox_to_anchor=(1.1,1.05))
plt.show()

for state in uniStates:
    honey = rawData[rawData["State"]==state].groupby("Year")["Value"]
    ssum = honey.sum()
    if ssum.sum() < Large and ssum.sum() > Larger:
        plt.plot(ssum.keys(),ssum,label=state)
plt.ticklabel_format(style='plain')
plt.xlabel("Year")
plt.ylabel("Production")
plt.title("Large Producers")
plt.legend(bbox_to_anchor=(1.1,1.05))
plt.show()

for state in uniStates:
    honey = rawData[rawData["State"]==state].groupby("Year")["Value"]
    ssum = honey.sum()
    if ssum.sum() < Larger and ssum.sum() > Medium:
        plt.plot(ssum.keys(),ssum,label=state)
plt.ticklabel_format(style='plain')
plt.xlabel("Year")
plt.ylabel("Production")
plt.title("Medium Producers")
plt.legend(bbox_to_anchor=(1.1,1.05))
plt.show()

for state in uniStates:
    honey = rawData[rawData["State"]==state].groupby("Year")["Value"]
    ssum = honey.sum()
    if ssum.sum() < Medium:
        plt.plot(ssum.keys(),ssum,label=state)
plt.ticklabel_format(style='plain')
plt.xlabel("Year")
plt.ylabel("Production")
plt.title("Lowest Producers")
plt.legend(bbox_to_anchor=(1.1,1.05))
plt.show()
```
