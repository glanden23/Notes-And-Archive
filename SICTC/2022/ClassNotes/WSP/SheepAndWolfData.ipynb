### FUNCTIONS

```python
def average(data):
    return sum(data)/len(data)

def splitValues(data):
    sheep = []
    wolf = []
    for i in data:
        sheep.append(float(i[1]))
        wolf.append(float(i[3]))
    return sheep, wolf

def aboutSame(data1,data2):
    sameCount = 0
    for i in range(len(data1)):
        if data1[i] - data2[i] < 10 and data1[i] - data2[i] > -10 or data2[i] - data1[i] < 10 and data2[i] - data1[i] > -10:
            sameCount+=1
    return sameCount

def separateData(data):
    sheep = []
    wolf = []
    for i in data:
        sheep.append([float(i[0]),float(i[1])])
        wolf.append([float(i[2]),float(i[3])])
    return sheep, wolf

def separateGraph(data):
    dataXs = []
    dataYs = []
    for i in data:
        dataXs.append(i[0])
        dataYs.append(i[1])
    return dataXs, dataYs

def loadAndClean(file):
    data = open(file).readlines()
    cleanData = []

    for i in data:
        cleanData.append(i.replace("\n","").split(","))
    return cleanData

```

```python
import matplotlib.pyplot as plt

data = loadAndClean("./data/noGrass/sheepTakeOver.csv")
sheep, wolf = separateData(data)
sheepVs, wolfVs = splitValues(data)

print("Sheeps Data\n","Min: "+str(min(sheepVs)),"Max: "+str(max(sheepVs)),"Average: "+str(average(sheepVs)))
print("Wolf Data\n","Min: "+str(min(wolfVs)),"Max: "+str(max(wolfVs)),"Average: "+str(average(wolfVs)))

print("Sheep and wolf counts are within 10 of each other",aboutSame(sheepVs,wolfVs),"times")

data1, data2 = separateGraph(sheep)
plt.plot(data1, data2, label="Sheeps")
data1, data2 = separateGraph(wolf)
plt.plot(data1, data2, label="Wolves")
plt.xlabel("Ticks")
plt.ylabel("Amount of Wolves/Sheep")
plt.title("Sheep Take Over")
plt.legend()
plt.show()
```

### FUNCTIONS FOR GRASS

```python
def average(data):
    return sum(data)/len(data)

def splitValues(data):
    sheep = []
    wolf = []
    grass = []
    for i in data:
        sheep.append(float(i[1]))
        wolf.append(float(i[3]))
        grass.append(float(i[5]))
    return sheep, wolf, grass

def aboutSame(data1,data2,data3):
    sameCount = 0
    for i in range(len(data1)):
        #https://stackoverflow.com/questions/21840165/check-if-variable-values-are-close-to-each-other
        if abs(data1[i] - data2[i]) <= 10 and abs(data2[i] - data3[i]) <= 10 and abs(data1[i] - data3[i]):
            sameCount+=1
    return sameCount

def separateData(data):
    sheep = []
    wolf = []
    grass = []
    for i in data:
        sheep.append([float(i[0]),float(i[1])])
        wolf.append([float(i[2]),float(i[3])])
        grass.append([float(i[4]),float(i[5])])
    return sheep, wolf, grass

def separateGraph(data):
    dataXs = []
    dataYs = []
    for i in data:
        dataXs.append(i[0])
        dataYs.append(i[1])
    return dataXs, dataYs

def loadAndClean(file):
    data = open(file).readlines()
    cleanData = []

    for i in data:
        cleanData.append(i.replace("\n","").split(","))
    return cleanData

```
